```python
weights = np.random.random(20)
weights /= weights.sum()
weights
```

```python
exp_return = np.sum(log_returns.mean()*weights)*255
```

```python
exp_volume = np.sqrt(np.dot(weights.T, np.dot(log_returns.cov()*255, weights)))
```

```python
sharpe_ratio1 = exp_return / exp_volume
```

```python
sharpe_ratio1
```

```python
n = 500

weights1 = np.zeros((n, 20))
exp_returns = np.zeros(n)
exp_volumes = np.zeros(n)
sharpe_ratios = np.zeros(n)

for i in range(n):
    weight = np.random.random(20)
    weight /= weight.sum()
    weights1[i] = weight
    
    exp_returns[i] = np.sum(log_returns.mean()*weight)*252
    exp_volumes[i] = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*255, weight)))
    sharpe_ratios[i] = exp_returns[i] / exp_volumes[i]
```

```python
sharpe_ratios.max()
```

```python
sharpe_ratios.argmax()
```

```python
fig, ax = plt.subplots()
ax.scatter(exp_volumes, exp_returns, c=sharpe_ratios)
ax.scatter(exp_volumes[sharpe_ratios.argmax()], exp_returns[sharpe_ratios.argmax()], c='r')
ax.set_xlabel('Expected Volatility')
ax.set_ylabel('Expected Return')
```
