Monte Carlo Simulation

```python
n=10000
weights = np.zeros((n,20))
exp_returns = np.zeros(n)
exp_vols = np.zeros(n)
sharpe_ratios = np.zeros(n)

for i in range (n):
    weight = np.random.random(20)
    weight /= weight.sum()
    weights[i] = weight

    exp_returns[i] = np.sum((log_returns.mean()*weight))*252
    exp_vols[i] = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252,weight)))
    sharpe_ratios[i] = (exp_returns[i]-(0.075/364))/exp_vols[i]
```

```python
sharpe_ratios.max()
```

```python
sharpe_ratios.argmax()
```

```python
# sharpe_ratios[2512]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(exp_vols,exp_returns,c=sharpe_ratios)
ax.scatter(exp_vols[sharpe_ratios.argmax()],exp_returns[sharpe_ratios.argmax()],c='red')
ax.set_xlabel('Expected Volatility')
ax.set_ylabel('Expected Returns')
```

```python
nifty50 = pdr.get_data_yahoo("^NSEI", start)
log_returns['Nifty50'] = np.log(nifty50['Adj Close']/nifty50['Adj Close'].shift())
log_returns.corr()
```

```python
fig1 = plt.figure(figsize=(20,20))
dataplot = sns.heatmap(log_returns.corr(),cmap='YlGnBu',annot= True)

plt.show(fig1)
```

```python
alphas = np.zeros(len(tickers_1))
betas = np.zeros(len(tickers_1))
def linear_regression(ticker_a,n):
    X = log_returns[ticker_a].to_numpy().reshape(-1, 1)
    Y = log_returns['Nifty50'].to_numpy().reshape(-1, 1)
    lin_regr = LinearRegression()
    lin_regr.fit(X, Y)
    Y_pred = lin_regr.predict(X)
    alpha = lin_regr.intercept_[0]
    beta = lin_regr.coef_[0, 0]
    alphas[n]=round(alpha, 5)
    betas[n]=round(beta, 3)
    fig, ax = plt.subplots()
    ax.set_title("Alpha: " + str(round(alpha, 5)) + ", Beta: " + str(round(beta, 3)))
    ax.scatter(X, Y)
    ax.plot(X, Y_pred, c='r')
```

```python
for i in range (len(tickers_1)):
    linear_regression(tickers_1[i],i)
```
